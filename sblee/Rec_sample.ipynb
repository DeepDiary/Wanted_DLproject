```python
from config.api_keys import gemini_key
import google.generativeai as genai
import pandas as pd
#from google.genai import types
```

```python
# CSV 파일 읽기
df = pd.read_csv('./data/ai_hub_data.csv')

# 감정 카테고리 매핑
emotion_mapping = {
    '분노': 'anger',
    '혐오': 'disgust',
    '공포': 'fear',
    '행복': 'happiness',
    '슬픔': 'sadness',
    '놀람': 'surprise',
    '중립': 'neutral'
}

# 감정 레이블 변환
df['Emotion'] = df['Emotion'].map(emotion_mapping)

# Gemini API 클라이언트 초기화
genai.configure(api_key=gemini_key)

# 시스템 명령어: 챗봇의 역할을 설정
# 이 명령은 챗봇이 어떤 종류의 응답을 생성해야 하는지 안내
system_instruction="당신은 사용자의 기분 전환을 돕는 추천 전문가입니다."

model = genai.GenerativeModel(
    'gemini-2.0-flash',
    system_instruction= system_instruction
    )

# Gemini 모델에 전달할 프롬프트를 생성
def create_prompt(emotion, conversation):
    prompt = f"""감정 '{emotion}'과 다음 대화를 기반으로 추천해 주세요:
대화: "{conversation}"
요청:
1. 기분을 개선하는 데 도움이 될 수 있는 세 가지 활동
2. 위로가 될 수 있는 저렴한 제품(50달러 미만) 세 가지
아래와 같은 형식으로 답변해 주세요:
활동:
1. [활동 1]
2. [활동 2]
3. [활동 3]
제품:
1. [제품 1]
2. [제품 2]
3. [제품 3]
"""
    return prompt

# Gemini 모델을 사용하여 주어진 감정과 대화에 대한 추천을 생성 
def get_recommendations(emotion, conversation):
    prompt = create_prompt(emotion, conversation)
    response = model.generate_content(prompt)
    return response.text

# 메인 함수: DataFrame의 각 행을 반복하고 감정과 대화를 추출하여 추천을 생성 
def main():
    for _, row in df.iterrows():
        emotion = row['Emotion']
        conversation = row['Sentence'] 
        
        # 감정과 대화를 사용하여 추천을 생성 
        recommendations = get_recommendations(emotion, conversation)
        
        print(f"Emotion: {emotion}")
        print(f"Conversation: {conversation}")
        print("Recommendations:")
        print(recommendations)
        print("\n" + "="*50 + "\n")

if __name__ == "__main__":
    main()
```
