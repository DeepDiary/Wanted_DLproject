### 디렉토리 내의 json 파일 확인

```python
import json
import os

def process_json_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                # 여기서 각 JSON 파일에 대한 처리를 수행합니다
                process_single_json(data, filename)

def process_single_json(data, filename):
    print(f"Processing file: {filename}")
    print(f"도서 제목: {data['metadata']['doc_name']}")
    print(f"저자: {data['metadata']['author']}")
    print(f"요약: {data['summary'][:100]}...")  # 요약의 처음 100자만 출력
    print("-" * 50)

# 사용 예:
json_directory = './data/도서요약/기타'
process_json_files(json_directory)
```

### 디렉토리 내의 일부 json 파일 확인

```python
# 인덱스 사용으로 찾기 
import json
import os

def process_indexed_json_files(directory, start_index, end_index):
    json_files = [f for f in os.listdir(directory) if f.endswith('.json')]
    selected_files = json_files[start_index:end_index]
    
    for filename in selected_files:
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            process_single_json(data, file_path)

# 사용 예:
directory = './data/도서요약/기타'
process_indexed_json_files(directory, 0, 16)  # 처음 16개의 JSON 파일만 처리
```

### 디렉토리 내의 있는 json 파일을 하나의 json 파일로 병합
* 책 제목을 기준으로 병합해서 요약내용 보기

```python
import json
import os

def merge_json_files(directory):
    merged_data = {}

    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                
                book_title = data['metadata']['doc_name']
                if book_title not in merged_data:
                    merged_data[book_title] = data
                    merged_data[book_title]['summaries'] = []
                
                merged_data[book_title]['summaries'].append(data['summary'])

    return merged_data

def save_merged_data(merged_data, output_file):
    with open(output_file, 'w', encoding='utf-8') as outfile:
        json.dump(merged_data, outfile, ensure_ascii=False, indent=2)

# 사용 예
directory = './data/도서요약/기타'
output_file = 'merge_test.json'

merged_data = merge_json_files(directory)
save_merged_data(merged_data, output_file)
print(f"병합된 데이터가 {output_file}에 저장되었습니다.")
```

#### 디렉토리 내 일부 파일만 병합

```python
import json
import os

def merge_indexed_json_files(directory, start_index, end_index, output_file):
    merged_data = None
    summaries = []

    json_files = [f for f in os.listdir(directory) if f.endswith('.json')]
    selected_files = json_files[start_index:end_index]

    for filename in selected_files:
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            
            if merged_data is None:
                merged_data = data
                merged_data['summaries'] = []
            
            summaries.append(data['summary'])

    if merged_data:
        merged_data['summaries'] = summaries
        del merged_data['summary']

        with open(output_file, 'w', encoding='utf-8') as outfile:
            json.dump(merged_data, outfile, ensure_ascii=False, indent=2)

        print(f"병합된 데이터가 {output_file}에 저장되었습니다.")
    else:
        print("처리할 JSON 파일을 찾지 못했습니다.")

# 사용 예:
directory = './data/도서요약/기타'
output_file = 'merge_독일어형태론.json'
merge_indexed_json_files(directory, 0, 16, output_file)  # 처음 16개의 JSON 파일만 처리
```

```python
import json
import nltk
from nltk.tokenize import sent_tokenize

# NLTK 데이터 다운로드 (처음 실행 시 한 번만 필요)
nltk.download('punkt')

def truncate_summary(summary, max_length=500):
    # 문장 단위로 분리
    sentences = sent_tokenize(summary)
    truncated_summary = ""
    current_length = 0

    for sentence in sentences:
        if current_length + len(sentence) <= max_length:
            truncated_summary += sentence + " "
            current_length += len(sentence) + 1
        else:
            break

    return truncated_summary.strip()

# JSON 파일 읽기
with open('merge_test.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 각 항목의 요약 수정
for key, value in data.items():
    if 'summaries' in value:
        value['summaries'] = [truncate_summary(summary) for summary in value['summaries']]

# 수정된 JSON 파일 저장
with open('merge_test1.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=2)

print("요약이 수정된 JSON 파일이 저장되었습니다.")
```

#### json > csv 파일로 저장

```python
import json
import csv

def json_to_csv(json_file, csv_file):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    with open(csv_file, 'w', newline='', encoding='utf-8') as f:
        csv_writer = csv.writer(f)

        # CSV 헤더 작성
        header = ['제목', '저자', '요약']
        csv_writer.writerow(header)

        # 데이터 작성
        for book_title, book_data in data.items():
            author = book_data['metadata']['author']
            summaries = '\n'.join(book_data['summaries'])  # 요약들을 개행문자로 연결
            row = [book_title, author, summaries]
            csv_writer.writerow(row)

    print(f"CSV 파일이 {csv_file}에 저장되었습니다.")

# 사용 예
json_file = './merge_test.json'
csv_file = './merge_test.csv'
json_to_csv(json_file, csv_file)
```
